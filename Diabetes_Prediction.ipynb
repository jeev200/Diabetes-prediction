Importing the Dependencies

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

```python
#loading the diabetes dataset to a pandas Dataframe
diabetes_dataset = pd.read_csv('diabetes1.csv',header=None)
```

```python
# Gives the first five rows of the dataset
diabetes_dataset.head()
```

```python
# Numbers of rows and columns in a dataset
diabetes_dataset.shape
```

```python
#Extract the first row to use as the new column names
new_columns = diabetes_dataset.iloc[0]
new_columns
```

```python
# Drop the first row from the DataFrame
diabetes_dataset = diabetes_dataset.drop(diabetes_dataset.index[0])
diabetes_dataset
```

```python
#Set the extracted row as the new column names
diabetes_dataset.columns = new_columns
diabetes_dataset
```

```python
#Reset the index
diabetes_dataset.reset_index(drop=True, inplace=True)
```

```python
# Convert columns to numeric types if they are not
diabetes_dataset = diabetes_dataset.apply(pd.to_numeric, errors='coerce')
```

```python
# Getting the statictical measures of the data
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

0 --> Non-Diabetic
1 --> Diabetic

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Separating the data and the labels
X = diabetes_dataset.drop(columns='Outcome',axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(Y)
```

# Data Standardzation

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
Standandarized_Data = scaler.transform(X)
```

```python
print(Standandarized_Data)
```

```python
X = Standandarized_Data
```

Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
print(X_train.shape,Y_train.shape,X_test.shape)
```

Training the model

```python
classifier = svm.SVC(kernel='linear')
```

```python
# Training the support vector machine classifier
classifier.fit(X_train,Y_train)
```

Model Evaluation

```python
# Accuracy Score on the training data
x_train_pred = classifier.predict(X_train)
training_data_accuracy = accuracy_score(x_train_pred,Y_train)
print('Accuracy on Training Data: ',training_data_accuracy)
```

```python
# Accuracy Score on the test data
x_test_pred = classifier.predict(X_test)
test_data_accuracy = accuracy_score(x_test_pred,Y_test)
print('Accuracy on test Data: ',test_data_accuracy)
```
